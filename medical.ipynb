```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# 1
df = pd.read_csv('medical_examination.csv')
df.shape
```

```python
# 2
df['overweight'] =  (df['weight'] / ((df['height'] / 100) ** 2)).apply(lambda x: 1 if x > 25 else 0)
df.head(5)
```

```python
# 3
df['cholesterol'] = df['cholesterol'].apply(lambda x: 1 if x > 1 else 0)
df['gluc'] = df['gluc'].apply(lambda x: 1 if x > 1 else 0)
df.head(5)
```

```python
# 5
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

```python
# 6
df_cat = df_cat.groupby(['cardio', 'variable', 'value'], as_index=False).size()
df_cat
```

```python
df_cat.rename(columns={'size': 'total'}, inplace=True)
df_cat
```

```python
# 7
fig = sns.catplot(x='variable', y='total', hue='value', col='cardio', kind='bar', data=df_cat)
```

```python
# 8
fig.savefig('catplot.png')
```

```python
# 11. Clean the data
df_heat = df[(df['ap_lo'] <= df['ap_hi']) & 
             (df['height'] >= df['height'].quantile(0.025)) &
             (df['height'] <= df['height'].quantile(0.975)) &
             (df['weight'] >= df['weight'].quantile(0.025)) &
             (df['weight'] <= df['weight'].quantile(0.975))]
df_heat
```

```python
# 12. Calculate the correlation matrix
corr = df_heat.corr()
corr
```

```python
# 13. Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```

```python
# 14. Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12, 8))
# 15. Draw the heatmap
sns.heatmap(corr, mask=mask, annot=True, fmt='.1f', cmap='coolwarm', square=True, linewidths=0.5, ax=ax)
plt.show()
```

```python
# 16. Save and return the figure
fig.savefig('heatmap.png')
```
